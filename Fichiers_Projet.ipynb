```python
import numpy as np
import pandas as pd
import os
import wget
import shutil
import gzip
import datetime
```

```python
date = []
for i in range (1996, 2022):
    for j in range(1, 13):
        annee = str(i)
        mois = str (j)
        if len(mois) == 1:
            mois = str(0)+mois
        date.append(annee+mois)
```

```python
start = "https://donneespubliques.meteofrance.fr/donnees_libres/Txt/Synop/Archive/synop."
end = ".csv.gz"

if not(os.path.isfile("Data_météo_R.csv")):
    for i in range (len(date)-2):
        lien_telechargement = start+date[i]+end
        file = date[i]+".csv.gz"
        wget.download(lien_telechargement, file)
        with gzip.open(file, 'rb') as f_in:
            with open(file[:-3], 'wb') as f_out:
                shutil.copyfileobj(f_in, f_out)

for i in range (len(date)-2):
    file = date[i]+".csv.gz"
    os.remove(file)
```

```python
telechargement = "https://donneespubliques.meteofrance.fr/donnees_libres/Txt/Synop/postesSynop.csv"
if not(os.path.isfile("postesMeteo.csv")):
    file = "postesMeteo.csv"
    wget.download(telechargement, file)
```

```python
final_dataframe = pd.read_csv('199601.csv')
os.remove('199601.csv')
```

```python
for i in range (1, len(date)-2):
    data_df = pd.read_csv(date[i]+'.csv')
    final_dataframe = pd.concat([final_dataframe, data_df]).reset_index(drop=True)

for i in range (1, len(date)-2):
        os.remove(date[i]+'.csv')
```

```python
final_dataframe.to_csv('Data_météo.csv', index=False) 
```
